## 1. 핵심 개념 정리

### 1) 스택 (Stack): 후입선출 (LIFO, Last-In First-Out)
- **개념**: 가장 마지막에 들어온 데이터가 가장 먼저 나가는 구조입니다.
- **특징**: 데이터의 삽입과 삭제가 한쪽 끝(**Top**)에서만 일어납니다.
- **대표 비유**: 프링글스 과자통, 차곡차곡 쌓인 접시
- **주요 연산**:
  - `push()`: 스택의 맨 위에 데이터를 추가 (파이썬 리스트의 `append()` 사용)
  - `pop()`: 스택의 맨 위 데이터를 제거하고 반환 (파이썬 리스트의 `pop()` 사용)
  - `peek()`: 맨 위 데이터를 꺼내지 않고 확인만 함 (파이썬 리스트의 `[-1]` 조회)

### 2) 큐 (Queue): 선입선출 (FIFO, First-In First-Out)
- **개념**: 먼저 들어온 데이터가 먼저 나가는 공평한 구조입니다.
- **특징**: 한쪽 끝(**Rear/Tail**)에서는 삽입만, 다른 쪽 끝(**Front/Head**)에서는 삭제만 일어납니다.
- **대표 비유**: 맛집 대기줄, 일방통행 터널
- **주요 연산**:
  - `enqueue()`: 큐의 맨 뒤에 데이터를 추가 (파이썬 deque의 `append()` 사용)
  - `dequeue()`: 큐의 맨 앞 데이터를 꺼내고 삭제 (파이썬 deque의 `popleft()` 사용)

### 3) 스택 vs 큐 비교 요약
| 구분 | 스택 (Stack) | 큐 (Queue) |
| :--- | :--- | :--- |
| **작동 원리** | 후입선출 (LIFO) | 선입선출 (FIFO) |
| **데이터 삽입** | 한쪽 끝 (Top) | 맨 뒤 (Rear) |
| **데이터 삭제** | 동일한 끝 (Top) | 맨 앞 (Front) |
| **추천 파이썬 모듈** | 기본 `list` | `collections.deque` |
| **실무 활용 사례** | 웹 브라우저 뒤로가기, Ctrl+Z(실행취소) | 인쇄 대기열, 은행 번호표 대기 |

## 2. 완성 코드 예제

아래의 완성된 코드를 실행하며 스택과 큐의 기본적인 동작 흐름을 파악해 보세요.

```python
# ==========================================
# 1. 파이썬 기본 리스트를 활용한 스택(Stack) 구현
# ==========================================
print("=== [예제] 스택(Stack) 동작 ===")
stack = []

# 데이터 삽입 (Push)
stack.append("A")
stack.append("B")
stack.append("C")
print("데이터 Push 완료:", stack)

# 데이터 확인 (Peek)
if stack:
    print("현재 맨 위의 데이터 (Peek):", stack[-1])

# 데이터 추출 (Pop)
pop_item = stack.pop()
print(f"Pop 연산 수행: {pop_item} 꺼냄")
print("Pop 이후 남은 스택:", stack)


# ==========================================
# 2. collections.deque를 활용한 고성능 큐(Queue) 구현
# ==========================================
print("\n=== [예제] 큐(Queue) 동작 ===")
from collections import deque

queue = deque()

# 데이터 삽입 (Enqueue)
queue.append("A")
queue.append("B")
queue.append("C")
print("데이터 Enqueue 완료:", queue)

# 데이터 추출 (Dequeue)
# 일반 리스트의 pop(0)은 O(N)의 시간이 걸리지만, deque의 popleft()는 O(1)로 매우 빠릅니다.
dequeue_item = queue.popleft()
print(f"Dequeue 연산 수행: {dequeue_item} 꺼냄")
print("Dequeue 이후 남은 큐:", queue)
```

## 5. 정답 및 풀이

작성한 코드와 아래의 정답 코드를 비교해 보세요.

```python
# ==========================================
# 1. 빈칸 채우기 정답
# ==========================================
# [빈칸 1]: append
# [빈칸 2]: pop
# [빈칸 3]: 0
# [빈칸 4]: deque
# [빈칸 5]: append
# [빈칸 6]: popleft


# ==========================================
# 2. 응용 문제 1 정답 (문자열 뒤집기)
# ==========================================
def reverse_string_sol(text):
    stack = []
    for char in text:
        stack.append(char) # Push
        
    reversed_text = ""
    while len(stack) > 0:
        reversed_text += stack.pop() # Pop
        
    return reversed_text

print("--- 응용 1 정답 확인 ---")
print(reverse_string_sol("HELLO")) # OLLEH


# ==========================================
# 3. 응용 문제 2 정답 (회문 판별기)
# ==========================================
from collections import deque

def is_palindrome_sol(word):
    word = word.replace(" ", "").lower()
    char_queue = deque(word)
    
    while len(char_queue) > 1:
        # 맨 앞(왼쪽)과 맨 뒤(오른쪽)에서 각각 하나씩 꺼내어 비교
        if char_queue.popleft() != char_queue.pop():
            return False
            
    return True

print("\n--- 응용 2 정답 확인 ---")
print("토마토:", is_palindrome_sol("토마토")) # True
print("Python:", is_palindrome_sol("Python")) # False
```
