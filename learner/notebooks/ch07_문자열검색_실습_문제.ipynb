# 1. 브루트 포스법 (Brute Force Method)


## 알고리즘 개념

**브루트 포스법**은 텍스트의 각 위치에서 패턴을 처음부터 하나씩 대조하는 가장 기본적인 문자열 검색 방법임.

- 구현이 단순하여 이해하기 쉬움
- 불일치 발생 시 이미 비교한 정보를 버리고 처음부터 다시 비교

### 핵심 변수
| 변수 | 역할 |
|------|------|
| `pt` | 텍스트(txt)를 따라가는 커서 |
| `pp` | 패턴(pat)을 따라가는 커서 |

### 커서 업데이트 규칙
- `txt[pt] == pat[pp]` → `pt += 1`, `pp += 1`
- `txt[pt] != pat[pp]` → `pt = pt - pp + 1`, `pp = 0`
- `pp == len(pat)` → 검색 성공, `return pt - pp`


## 1. 코드 읽기

아래 완성된 코드를 실행해 보고 동작을 이해해 봅시다.


````python
# 1단계: 완성된 브루트 포스법 코드를 읽고 동작을 이해해 봅시다.

def bf_match(txt: str, pat: str) -> int:
    """브루트 포스법으로 문자열 검색"""
    pt = 0   # txt를 따라가는 커서
    pp = 0   # pat를 따라가는 커서

    while pt != len(txt) and pp != len(pat):
        if txt[pt] == pat[pp]:
            pt += 1
            pp += 1
        else:
            pt = pt - pp + 1   # 비교 시작 위치의 다음 칸
            pp = 0             # 패턴 처음부터 다시

    return pt - pp if pp == len(pat) else -1

# 동작 확인
s1 = "ABABCDEFGHA"
s2 = "ABC"

idx = bf_match(s1, s2)
if idx == -1:
    print("텍스트 안에 패턴이 존재하지 않습니다.")
else:
    print(f"{idx + 1}번째 문자가 일치합니다.")
````

## 3. 실행 과정 추적

**확인 문제**  
`txt = "ABABCDEFGHA"`, `pat = "ABC"` 일 때:
1. 몇 번의 시도(시작 위치)를 거쳐 검색에 성공하는가?
2. 최종 반환값 `pt - pp`는 무엇인가?
3. 출력값(`idx + 1`)은 무엇인가?

아래 셀을 실행하여 단계별 추적 결과를 확인하세요.


````python
# 브루트 포스법 단계별 실행 추적

def bf_match_verbose(txt: str, pat: str) -> int:
    pt = 0
    pp = 0
    step = 0

    print(f"텍스트: {txt}")
    print(f"패턴:   {pat}")
    print("-" * 50)

    while pt != len(txt) and pp != len(pat):
        step += 1
        match = txt[pt] == pat[pp]
        status = "일치" if match else "불일치"
        print(f"[단계 {step:2d}] pt={pt}, pp={pp} | txt[{pt}]='{txt[pt]}', pat[{pp}]='{pat[pp]}' → {status}")

        if match:
            pt += 1
            pp += 1
        else:
            print(f"         → 시작 위치 {pt-pp} → {pt-pp+1} 로 이동, pp 초기화")
            pt = pt - pp + 1
            pp = 0

    result = pt - pp if pp == len(pat) else -1
    print("-" * 50)
    if result != -1:
        print(f"검색 성공! 시작 인덱스(Python): {result}  →  {result+1}번째 문자")
    else:
        print("검색 실패: 패턴이 존재하지 않습니다.")
    return result

bf_match_verbose("ABABCDEFGHA", "ABC")
````

# 2. KMP법 (Knuth-Morris-Pratt)


## 알고리즘 개념

**KMP법**은 불일치 발생 시 이미 비교한 결과를 버리지 않고, **건너뛰기 표(skip 배열)**를 이용해 효율적으로 재활용하는 알고리즘임.

- 텍스트 커서 `pt`는 절대 뒤로 이동하지 않음
- 시간 복잡도: O(n) (최악의 경우에도)

### KMP 두 단계

| 단계 | 목적 | 핵심 동작 |
|------|------|-----------|
| 1단계: skip 만들기 | 패턴의 반복 구조 분석 | 앞부분과 뒷부분이 같은 최대 길이 저장 |
| 2단계: 검색하기 | skip을 활용한 검색 | 불일치 시 `pp = skip[pp]`로 되돌림 |

### skip 배열의 의미
- `skip[n]` = "패턴의 앞 n글자를 봤을 때, 앞부분과 뒷부분이 같은 최대 길이"
- 패턴 `ABCABD`: `skip = [0, 0, 0, 0, 1, 2, 0]`
  - `skip[5] = 2` → `ABCAB`에서 앞 `AB`와 뒤 `AB`가 일치 (길이 2)


## 1. 코드 읽기

아래 완성된 코드를 실행해 보고 동작을 이해해 봅시다.


````python
# 1단계: 완성된 KMP법 코드를 읽고 동작을 이해해 봅시다.

def kmp_match(txt: str, pat: str) -> int:
    """KMP법으로 문자열 검색"""
    pt = 1   # 패턴 분석용 커서 (1부터 시작)
    pp = 0   # 패턴 분석용 커서
    skip = [0] * (len(pat) + 1)  # skip 배열: 인덱스 0~len(pat)

    # 1단계: skip 배열 만들기 (패턴 자신과 비교)
    skip[pt] = 0
    while pt != len(pat):
        if pat[pt] == pat[pp]:
            pt += 1
            pp += 1
            skip[pt] = pp        # 앞뒤가 pp글자 일치
        elif pp == 0:
            pt += 1
            skip[pt] = 0         # 겹침 없음
        else:
            pp = skip[pp]        # 더 짧은 후보로 이동

    # 2단계: 검색하기
    pt = pp = 0
    while pt != len(txt) and pp != len(pat):
        if txt[pt] == pat[pp]:
            pt += 1
            pp += 1
        elif pp == 0:
            pt += 1              # 첫 글자부터 틀림: txt만 이동
        else:
            pp = skip[pp]        # skip으로 pp 되돌리기

    return pt - pp if pp == len(pat) else -1

# 동작 확인
s1 = "ZABCABXABCABD"
s2 = "ABCABD"
idx = kmp_match(s1, s2)
if idx == -1:
    print("텍스트 안에 패턴이 존재하지 않습니다.")
else:
    print(f"{idx + 1}번째 문자가 일치합니다.")  # 기대: 8번째
````

## 2. skip 배열 이해하기

skip 배열을 직접 출력하여 각 값의 의미를 확인해 봅시다.


````python
# skip 배열을 시각적으로 확인하기

def make_skip(pat: str) -> list:
    pt = 1
    pp = 0
    skip = [0] * (len(pat) + 1)
    skip[pt] = 0
    while pt != len(pat):
        if pat[pt] == pat[pp]:
            pt += 1; pp += 1; skip[pt] = pp
        elif pp == 0:
            pt += 1; skip[pt] = 0
        else:
            pp = skip[pp]
    return skip

pat = "ABCABD"
skip = make_skip(pat)

print(f"패턴: {pat}")
print(f"인덱스: {list(range(len(skip)))}")
print(f"skip:   {skip}")
print()
print("skip[n]의 의미: '패턴 앞 n글자를 봤을 때, 앞부분=뒷부분인 최대 길이'")
print(f"skip[4]=1: 'ABCA' → 앞 'A' == 뒤 'A' (길이 1)")
print(f"skip[5]=2: 'ABCAB' → 앞 'AB' == 뒤 'AB' (길이 2)")
print(f"skip[6]=0: 'ABCABD' → 앞뒤 공통 부분 없음")
````

## 4. 실행 과정 추적

**확인 문제**  
`txt = "ZABCABXABCABD"`, `pat = "ABCABD"` 일 때:
1. `pt=6, pp=5`(txt[6]='X', pat[5]='D') 에서 불일치가 발생했을 때 pp는 어떻게 변하는가?
2. 최종 검색 성공 위치(1-based)는 몇 번째인가?

아래 셀을 실행하여 단계별 추적 결과를 확인하세요.


````python
# KMP법 단계별 실행 추적

def kmp_match_verbose(txt: str, pat: str) -> int:
    pt = 1; pp = 0
    skip = [0] * (len(pat) + 1)
    skip[pt] = 0
    while pt != len(pat):
        if pat[pt] == pat[pp]: pt+=1; pp+=1; skip[pt]=pp
        elif pp == 0: pt+=1; skip[pt]=0
        else: pp=skip[pp]

    print(f"텍스트: {txt}")
    print(f"패턴:   {pat}")
    print(f"skip:   {skip}")
    print("-"*60)

    pt = pp = 0; step = 0
    while pt != len(txt) and pp != len(pat):
        step += 1
        match = txt[pt] == pat[pp]
        status = "일치" if match else "불일치"
        print(f"[단계{step:2d}] pt={pt:2d},pp={pp} | txt[{pt}]='{txt[pt]}',pat[{pp}]='{pat[pp]}' → {status}", end="")
        if match:
            pt+=1; pp+=1; print()
        elif pp==0:
            print(f" (pp==0 → pt+1)"); pt+=1
        else:
            old=pp; pp=skip[pp]
            print(f" (pp: {old}→skip[{old}]={pp})")

    result = pt-pp if pp==len(pat) else -1
    print("-"*60)
    if result!=-1:
        print(f"검색 성공! 인덱스(Python)={result} → {result+1}번째 문자")
    else:
        print("검색 실패")
    return result

kmp_match_verbose("ZABCABXABCABD", "ABCABD")
````

# 3. 보이어·무어법 (Boyer-Moore Method)


## 알고리즘 개념

**보이어·무어법**은 패턴의 **오른쪽 끝 문자부터 역방향으로 비교**하고, 불일치 시 skip 표를 이용해 패턴을 크게 이동시키는 알고리즘임.

- 평균 시간 복잡도: O(n/m) — 실제로 가장 빠른 문자열 검색 알고리즘
- 실제 텍스트 편집기, 검색 엔진에서 널리 사용

### skip 표 계산 규칙
```
skip[문자] = 패턴 길이 n          (패턴에 없는 문자)
skip[pat[k]] = n - k - 1         (패턴에 있는 문자, 마지막 등장 위치 k 기준)
```

### 이동량 계산
```python
pt += max(skip[ord(txt[pt])], len(pat) - pp)
# move1 = skip[ord(txt[pt])]: 불일치 문자 기준 이동량
# move2 = len(pat) - pp:      최소 이동량 보정 (0칸 이동 방지)
```


## 1. 코드 읽기

아래 완성된 코드를 실행해 보고 동작을 이해해 봅시다.


````python
# 1단계: 완성된 보이어·무어법 코드를 읽고 동작을 이해해 봅시다.

def bm_match(txt: str, pat: str) -> int:
    """보이어·무어법으로 문자열 검색"""
    skip = [None] * 256   # ASCII 256개 문자의 이동량

    # skip 표 만들기
    for pt in range(256):
        skip[pt] = len(pat)                      # 기본값: 패턴에 없는 문자는 n칸 이동
    for pt in range(len(pat)):
        skip[ord(pat[pt])] = len(pat) - pt - 1   # 패턴 문자: 끝에서의 거리

    # 검색하기 (패턴 끝 문자부터 역방향 비교)
    while pt < len(txt):
        pp = len(pat) - 1          # 패턴 커서를 맨 뒤로
        while txt[pt] == pat[pp]:  # 일치하면 왼쪽으로 이동
            if pp == 0:
                return pt          # 패턴 전체 일치 → 시작 위치 반환
            pt -= 1
            pp -= 1
        # 불일치: move1과 move2 중 큰 값만큼 이동
        pt += skip[ord(txt[pt])] if skip[ord(txt[pt])] > len(pat) - pp else len(pat) - pp

    return -1

# 동작 확인
s1 = "ABABCDEFGHA"
s2 = "ABC"
idx = bm_match(txt=s1, pat=s2)
if idx == -1:
    print("텍스트 안에 패턴이 존재하지 않습니다.")
else:
    print(f"{idx + 1}번째 문자가 일치합니다.")  # 기대: 3번째
````

## 2. skip 표 이해하기

skip 표를 직접 출력하여 각 값의 의미를 확인해 봅시다.


````python
# skip 표를 시각적으로 확인하기

def show_skip(pat: str):
    skip = [len(pat)] * 256
    for pt in range(len(pat)):
        skip[ord(pat[pt])] = len(pat) - pt - 1

    print(f"패턴: {pat}  (길이 n={len(pat)})")
    print(f"{'문자':<6} {'ord':<5} {'skip 값':<8} 계산식")
    print("-" * 40)
    for c in sorted(set(pat)):
        k = pat.rfind(c)
        val = skip[ord(c)]
        print(f"'{c}'    {ord(c):<5} {val:<8} n-{k}-1 = {len(pat)}-{k}-1 = {val}")
    print(f"그 외  -      {len(pat):<8} 패턴에 없는 문자 → n칸 이동")

show_skip("ABC")
print()
show_skip("ABCA")  # A가 두 번 등장 → 마지막 위치 기준으로 덮어씀
````

아래 코드의 빈칸(`_____`)을 채워 보이어·무어법을 완성하세요.

## 4. 실행 과정 추적

**확인 문제**  
`txt = "ABABCDEFGHA"`, `pat = "ABC"` 일 때:
1. skip 표에서 A, B, C, 그 외 문자의 이동량은?
2. 첫 번째 비교(pt=2, pp=2)에서 불일치 발생 시 pt는 몇으로 이동하는가?
3. 검색 성공까지 총 몇 번의 비교가 이루어지는가?

아래 셀을 실행하여 단계별 추적 결과를 확인하세요.


````python
# 보이어·무어법 단계별 실행 추적

def bm_match_verbose(txt: str, pat: str) -> int:
    skip = [len(pat)] * 256
    for pt in range(len(pat)):
        skip[ord(pat[pt])] = len(pat) - pt - 1

    print(f"텍스트: {txt}")
    print(f"패턴:   {pat}")
    skip_display = {c: skip[ord(c)] for c in set(pat)}
    skip_display['그 외'] = len(pat)
    print(f"skip 표: {skip_display}")
    print("-" * 60)

    step = 0
    while pt < len(txt):
        pp = len(pat) - 1
        compare_pos = pt
        inner_steps = []

        while txt[pt] == pat[pp]:
            inner_steps.append(f"txt[{pt}]='{txt[pt]}'==pat[{pp}]='{pat[pp]}'✓")
            if pp == 0:
                step += 1
                for s in inner_steps:
                    print(f"[단계{step:2d}] {s}")
                print(f"★ 검색 성공! pt={pt} → {pt+1}번째 문자")
                return pt
            pt -= 1; pp -= 1

        step += 1
        move1 = skip[ord(txt[pt])]
        move2 = len(pat) - pp
        chosen = move1 if move1 > move2 else move2
        print(f"[단계{step:2d}] 패턴 끝({pat[-1]}) → 왼쪽 비교 시작, pt={compare_pos}")
        for s in inner_steps:
            print(f"        {s}")
        print(f"        txt[{pt}]='{txt[pt]}'≠pat[{pp}]='{pat[pp]}' → 불일치")
        print(f"        move1(skip['{txt[pt]}'])={move1}, move2(n-pp={len(pat)}-{pp})={move2} → max={chosen}칸 이동")
        pt += chosen

    print("검색 실패")
    return -1

bm_match_verbose("ABABCDEFGHA", "ABC")
````

## 5. 응용 문제

**문제**: 보이어–무어법에서 패턴에 **같은 문자가 두 번 이상** 등장할 때 skip 표는 어떻게 설정되는지 확인하고,  
패턴 `"ABCA"`에 대한 skip 표를 직접 채우세요.

- `ord()` 사용 이유: 리스트는 숫자 인덱스만 허용하므로 문자를 `ord()`로 정수로 변환
- `for` 루프를 통해 같은 문자가 여러 번 등장하면 **마지막 등장 위치 기준**으로 덮어씀


````python
# 응용 문제: 중복 문자가 있는 패턴의 skip 표 완성하기

# pat = "ABCA" 의 skip 표를 손으로 계산한 뒤, 코드로 확인하세요.
# 계산:
#   pt=0: pat[0]='A' → skip['A'] = 4 - 0 - 1 = ?
#   pt=1: pat[1]='B' → skip['B'] = 4 - 1 - 1 = ?
#   pt=2: pat[2]='C' → skip['C'] = 4 - 2 - 1 = ?
#   pt=3: pat[3]='A' → skip['A'] = 4 - 3 - 1 = ?  ← 앞의 A 값을 덮어씀!

def show_skip_detail(pat: str):
    skip = [len(pat)] * 256
    print(f"패턴: '{pat}'  (n={len(pat)})")
    print("계산 과정:")
    for pt in range(len(pat)):
        old = skip[ord(pat[pt])]
        skip[ord(pat[pt])] = len(pat) - pt - 1
        new = skip[ord(pat[pt])]
        overwrite = " ← 덮어씀!" if old != len(pat) and old != new else ""
        print(f"  pt={pt}: '{pat[pt]}' → skip['{pat[pt]}'] = {new}{overwrite}")

    print(f"최종 skip 표:")
    for c in sorted(set(pat)):
        print(f"  '{c}': {skip[ord(c)]}")
    print(f"  그 외: {len(pat)}")

    print(f"주의: 실제 이동은 max(skip값, len(pat)-pp)이므로")
    print(f"skip['A']=0이어도 len(pat)-pp >= 1 이면 최소 1칸 이동이 보장됨")

show_skip_detail("ABCA")
````
